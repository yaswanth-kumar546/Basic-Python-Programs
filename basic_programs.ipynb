**Write a python program to slove the quadratic quation?**


```python
import math

# Input coefficients
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Calculate discriminant
d = b**2 - 4*a*c

if d > 0:
    root1 = (-b + math.sqrt(d)) / (2*a)
    root2 = (-b - math.sqrt(d)) / (2*a)
    print("Two real roots:", root1, "and", root2)
elif d == 0:
    root = -b / (2*a)
    print("One real root:", root)
else:
    real = -b / (2*a)
    imag = math.sqrt(-d) / (2*a)
    print("Complex roots:", real, "+", imag, "i  and", real, "-", imag, "i")
```

**write a program to check leaf year or not?**

```python
year = int(input("Enter a year: "))

# Leap year conditions:
# 1. Divisible by 4
# 2. Not divisible by 100, unless also divisible by 400

if (year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

**Write a program to check amstrong number?**
An Armstrong number (also called a narcissistic number) is a number that is equal to the sum of its own digits each raised to the power of the number of digits. For example, 153 is an Armstrong number because
1
3
+
5
3
+
3
3
=
153
.

```python
num = int(input("Enter a number: "))
sum_of_powers = 0
digits = len(str(num))

temp = num
while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** digits
    temp //= 10

if num == sum_of_powers:
    print(num, "is an Armstrong number.")
else:
    print(num, "is not an Armstrong number.")
```

**Write a program to find a armstrong number in interval?**

```python
# Program to find Armstrong numbers in an interval

start = int(input("Enter start of interval: "))
end = int(input("Enter end of interval: "))

print("Armstrong numbers between", start, "and", end, "are:")

for num in range(start, end + 1):
    digits = len(str(num))
    sum_of_powers = sum(int(digit) ** digits for digit in str(num))
    if num == sum_of_powers:
        print(num)
```

**Write a program to find fctorial of a number using recursion?**

```python
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    else:
        return n * factorial(n - 1)   # recursive call

num = int(input("Enter a number: "))
print("Factorial of", num, "is:", factorial(num))
```

**Write a program cube sum of prime numbers?**

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

n = int(input("Enter a number: "))
cube_sum = sum(i**3 for i in range(2, n+1) if is_prime(i))

print("Cube sum of prime numbers up to", n, "is:", cube_sum)
```

**Write a program for array rotation?**

```python
def rotate_array(arr, d):
    d = d % len(arr)   # handle rotations greater than length
    return arr[d:] + arr[:d]

# Example usage
arr = [1, 2, 3, 4, 5]
d = int(input("Enter number of rotations: "))

rotated = rotate_array(arr, d)
print("Array after rotation:", rotated)
```

**Write a program to check if given array is monotonic?**
Monotonic means something that moves in only one direction—either always increasing or always decreasing—without reversing. In mathematics, a monotonic function or sequence is one that is entirely non‑decreasing or non‑increasing across its domain.

```python
def is_monotonic(arr):
    return all(arr[i] <= arr[i+1] for i in range(len(arr)-1)) or \
           all(arr[i] >= arr[i+1] for i in range(len(arr)-1))

# Example usage
arr = [1, 2, 2, 3]
print("Array is monotonic:", is_monotonic(arr))

arr2 = [6, 5, 4, 4]
print("Array is monotonic:", is_monotonic(arr2))

arr3 = [1, 3, 2]
print("Array is monotonic:", is_monotonic(arr3))
```

**Write a program to add any two matrices?**

```python
# Example matrices
A = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

B = [[9, 8, 7],
     [6, 5, 4],
     [3, 2, 1]]

# Result matrix with same dimensions
result = [[0 for _ in range(len(A[0]))] for _ in range(len(A))]

# Matrix addition
for i in range(len(A)):
    for j in range(len(A[0])):
        result[i][j] = A[i][j] + B[i][j]

# Display result
print("Resultant Matrix after addition:")
for row in result:
    print(row)
```

**Write a program to multiply any two matrices?**

```python
# Example matrices
A = [[1, 2, 3],
     [4, 5, 6]]

B = [[7, 8],
     [9, 10],
     [11, 12]]

# Result matrix with dimensions (rows of A × columns of B)
result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

# Matrix multiplication
for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            result[i][j] += A[i][k] * B[k][j]

# Display result
print("Resultant Matrix after multiplication:")
for row in result:
    print(row)
```

**Write a program to transpose the matrix?**

```python
# Example matrix
A = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

# Transpose using list comprehension
transpose = [[A[j][i] for j in range(len(A))] for i in range(len(A[0]))]

# Display result
print("Original Matrix:")
for row in A:
    print(row)

print("\nTransposed Matrix:")
for row in transpose:
    print(row)
```
