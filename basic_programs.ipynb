**Write a calculator program in detailed?**

```python
# Take input from the user
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

# Perform calculation
if operator == "+":
    print("Result:", num1 + num2)
elif operator == "-":
    print("Result:", num1 - num2)
elif operator == "*":
    print("Result:", num1 * num2)
elif operator == "/":
    if num2 == 0:
        print("Error! Division by zero.")
    else:
        print("Result:", num1 / num2)
else:
    print("Invalid operator!")
```

**Write a program to check leaf year?**

```python
year = int(input("Enter a year: "))

# Leap year conditions:
# 1. Divisible by 4
# 2. Not divisible by 100, unless also divisible by 400

if (year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

**Write a program to check prime number?**

```python
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number.")
            break
    else:
        print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

**Write a prime checking number in less lines?**

```python
num = int(input("Enter a number: "))
if num > 1 and all(num % i != 0 for i in range(2, int(num**0.5)+1)):
    print("Prime")
else:
    print("Not Prime")
```

**Write a prime checking program in one line?**

```python
num = int(input("Enter the number: "))
print("Prime" if num > 1 and all(num % i != 0 for i in range(2, int(num**0.5)+1)) else "Not Prime")
```

**Write a program to convert celsius to Fahrenheit?**

```python
# Convertion Formula = Fahrenheit = (Celsius * 9/5) + 32

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32    # Fahrenheit is temperature scale, invented by Daniel Gabriel Fahrenheit.
print("Temperature in Fahrenheit:", fahrenheit)
```
